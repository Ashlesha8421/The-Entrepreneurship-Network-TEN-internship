What is image ?
* An image is a picture that has been created by **combination of vectors or values.**
* An image can be describe in terms of **vector graphic or raster graphics.**
* An image stored in raster form is sometimes called a **bitmap**

```python
# OpenCV bydefault read BGR (Blue,Green,Red)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.imshow(img_cv)
```

```python
#Rendering matplotlib array
plt.imshow(img_mat) #read image in (RBG format)
```

```python
#Analyzing image by matplotlib
#split array into three parts
r = img_mat[:,:,0] # red array
g = img_mat[:,:,1] # green array
b = img_mat[:,:,2] # Blue array
```

```python
r.shape,g.shape,b.shape
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,3,1)
plt.title('Red region')
plt.imshow(r, cmap='gray')
plt.subplot(1,3,2)
plt.title('Green region')
plt.imshow(g, cmap='gray')
plt.subplot(1,3,3)
plt.title('Blue region')
plt.imshow(b, cmap='gray')
plt.show()
```

```python
# Combination of red creen and blue 
# GrayScale_Image = 0.3*Red array + 0.59*Green array +0.11*Blue array
# Gray scale : 2d which consist of all the parts of information
# The reason is working with greay scale is that olor image sends same information in three matrix and same inforamtion which has a greayscale is a two dimensional.
# While combining the image,we see that the computaion ofcolr image is 3 times greater than gray color image which is shorter 3 time
# And Computationally greayscale are more and more efficient


gray_MAT = cv2.cvtColor(img_mat,cv2.COLOR_RGB2GRAY)
```

```python
# Gray scale :2d
gray_CV = cv2.cvtColor(img_cv,cv2.COLOR_BGR2GRAY)
```

```python
gray_MAT.shape,gray_CV.shape
```

```python
plt.imshow(gray_MAT,cmap='gray') # with matplotlib
```
