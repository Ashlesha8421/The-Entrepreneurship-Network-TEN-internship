```python
import numpy as np # numerical python
import cv2 # computer vision
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading image
img = cv2.imread('./Data/test.jpg') #bgr
#convert into grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(gray.shape)
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
#shrink (w=1600,h=1200)
#(w=400,h=300)
img_re =cv2.resize(gray,(500,300),cv2.INTER_AREA)
img_re.shape
```

```python
plt.imshow(img_re,cmap='gray')
plt.show()
```

```python
# enlarge (w=1000,h=1600)
#cv2.INTERCUBIC --> SLOW & GOOD
#cv2.INTERAREA --> FAST & OKAY
img_en =cv2.resize(gray,(2000,1600),cv2.INTER_CUBIC)
img_en.shape
```

```python
plt.imshow(img_en,cmap='gray')
plt.show()
```
